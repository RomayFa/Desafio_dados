### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
import matplotlib.pyplot as plt
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
print("Faturamento Loja 1:", faturamento_loja1)
print("Faturamento Loja 2:", faturamento_loja2)
print("Faturamento Loja 3:", faturamento_loja3)
print("Faturamento Loja 4:", faturamento_loja4)
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title('Faturamento Total por Loja', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.show()
```

# 2. Vendas por Categoria


```python
categoria_loja1 = loja['Categoria do Produto'].value_counts()
categoria_loja2 = loja2['Categoria do Produto'].value_counts()
categoria_loja3 = loja3['Categoria do Produto'].value_counts()
categoria_loja4 = loja4['Categoria do Produto'].value_counts()

tabela_categorias = pd.DataFrame({
    'Loja 1': categoria_loja1,
    'Loja 2': categoria_loja2,
    'Loja 3': categoria_loja3,
    'Loja 4': categoria_loja4
}).fillna(0).astype(int)
print("📋 Tabela de produtos vendidos por categoria em cada loja:")
print(tabela_categorias)
```

```python
cat1 = loja['Categoria do Produto'].value_counts()
cat2 = loja2['Categoria do Produto'].value_counts()
cat3 = loja3['Categoria do Produto'].value_counts()
cat4 = loja4['Categoria do Produto'].value_counts()
print("📊 Categoria mais vendida em cada loja:")
print(f"Loja 1: {cat1.idxmax()} ({cat1.max()} vendas)")
print(f"Loja 2: {cat2.idxmax()} ({cat2.max()} vendas)")
print(f"Loja 3: {cat3.idxmax()} ({cat3.max()} vendas)")
print(f"Loja 4: {cat4.idxmax()} ({cat4.max()} vendas)")
```

```python
labels = [
    f"Loja 1 ({cat1.idxmax()})",
    f"Loja 2 ({cat2.idxmax()})",
    f"Loja 3 ({cat3.idxmax()})",
    f"Loja 4 ({cat4.idxmax()})"
]

valores = [cat1.max(), cat2.max(), cat3.max(), cat4.max()]


plt.figure(figsize=(8, 6))
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('🔝 Categoria mais vendida em cada loja')
plt.axis('equal')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_loja1 = loja['Produto'].value_counts()
vendas_loja2 = loja2['Produto'].value_counts()
vendas_loja3 = loja3['Produto'].value_counts()
vendas_loja4 = loja4['Produto'].value_counts()

```

```python
def pegar_mais_menos_vendidos(vendas):
    mais_vendido = vendas.idxmax()
    menos_vendido = vendas.idxmin()
    return {mais_vendido: vendas[mais_vendido], menos_vendido: vendas[menos_vendido]}


def plot_mais_menos_vendidos(vendas, titulo):
    vendas_filtradas = pegar_mais_menos_vendidos(vendas)
    plt.figure(figsize=(6,4))
    plt.bar(vendas_filtradas.keys(), vendas_filtradas.values(), color=['green', 'red'])
    plt.title(titulo)
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Gráficos
plot_mais_menos_vendidos(vendas_loja1, "Mais e Menos Vendido - Loja 1")
plot_mais_menos_vendidos(vendas_loja2, "Mais e Menos Vendido - Loja 2")
plot_mais_menos_vendidos(vendas_loja3, "Mais e Menos Vendido - Loja 3")
plot_mais_menos_vendidos(vendas_loja4, "Mais e Menos Vendido - Loja 4")
```

# 5. Frete Médio por Loja


```python
frete_medio = {
    'Loja 1': loja['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}


lojas = list(frete_medio.keys())
custos = list(frete_medio.values())


plt.figure(figsize=(8,5))
plt.scatter(lojas, custos, color='purple', s=150)
plt.title('Custo Médio de Frete por Loja')
plt.xlabel('Lojas')
plt.ylabel('Custo Médio (R$)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
